```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Team': ['India', 'Pakisthan', 'England', 'South Africa', 'Australia','New Zealand','Afghanisthan'],
    'Matches Played': [4, 4, 4, 4, 4,5,4],
    'Wins': [3, 2, 2, 1, 0,3,1],
    'Draws': [0, 1, 0, 1, 2,0,2],
    'Losses': [1, 1, 2, 2, 2,2,1],
    'Points': [9, 7, 6, 4, 2,6,4]
}

df = pd.DataFrame(data)

# Sort the DataFrame by points in descending order
df = df.sort_values(by='Points', ascending=False)
print(df)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
colors = sns.color_palette("deep", len(df))

barplot = sns.barplot(x='Points', y='Team', data=df, palette=colors)
plt.title('Cricket World Cup Points Table')
plt.xlabel('Points')
plt.ylabel('Team')

for index, row in df.iterrows():
    barplot.text(row['Points'], index, f' {row["Points"]} pts', color='black', ha="left")

plt.show()
```
