```python
import numpy as np
import pandas as pd
```

```python
# Load the dataset from the CSV file
```

```python
data = pd.read_csv('mydataset.csv')
```

```python
print(data)
```

```python
# Remove missing values: Use the dropna() method to remove any row with at least one missing value.
data.dropna()
```

```python
# Fill missing values in 'Age' with mean

data['Age'].mean()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data
```

```python
# Fill missing values in 'Gender' with mode

data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
```

```python
data
```

```python
# Fill missing values in 'Height' and 'Weight' with mean

data['Height'] = data['Height'].fillna(data['Height'].mean())
data['Weight'] = data['Weight'].fillna(data['Weight'].mean())
```

```python
data
```

```python


data = data.drop_duplicates()
```

```python
data
```
